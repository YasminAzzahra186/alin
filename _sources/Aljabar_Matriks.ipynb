#Penjelasan Mengenai Matriks
##Pengertian Matriks:
Matriks adalah susunan bilagan real(atau elemen-elemen) yang disusun dalam baris dan kolom sehingga membentuk persegi ataupun persegi panjang. Setiap elemen memiliki posisi unik dalam matriks, ditentukan oleh baris dan kolomnya. Misalnya, matriks 2x3 memiliki 2 baris dan 3 kolom. suatu matriks biasa diberi nama dengan menggunakan huruf kapital (seperti A, B dll) dan elemennya dinyatakan dengan huruf kecil.


### Berikut Contoh Matriks
\begin{align*}
A=
\begin{bmatrix}
2&3&2\\

-1&-2&2
\end{bmatrix}
&&
B=
\begin{bmatrix}

5&-2&7
\end{bmatrix}
&&
C=

\end{bmatrix}
\end{align*}


#Operasi pada Matriks
Operasi matriks adalah serangkaian operasi atau tindakan yang dilakukan pada matriks, baik itu untuk mengubah matriks ataupun menggabungkan matriks. Berikut beberapa operasi matriks yang sering digunakan:

```python
import numpy as np

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])
c=np.array([[3,4],[-2,4],[2,5]])

print('A=\n',a)
print('B=\n',b)
print('C=\n',c)
```

##1.Penjumlahan matriks :

Operasi ini melibatkan penambahan elemen-elemen yang berada pada posisi yang sama dari dua matriks yang memiliki ordo yang sama. Artinya kedua matriks ini harus memiliki baris dan kolom yang sama.

CONTOH PENJUMLAHAN MATRIKS A + B :
\begin{align*}
A=
\begin{bmatrix}
2&3&2\\
4&1&-3\\
-1&-2&2
\end{bmatrix}
&&
B=
\begin{bmatrix}
2&4&7\\
6&-3&2\\
5&-2&7
\end{bmatrix}
\end{align*}

\begin{align*}
A + B =
\begin{bmatrix}
2+2&3+4&2+7\\
4+6&1+(-3)&-3+2\\
-1+5&-2+(-2)&2+7
\end{bmatrix}
=
\begin{bmatrix}
4&7&9\\
10&-2&-1\\
4&-4&9
\end{bmatrix}
\end{align*}

BERIKUT CODE PENJUMLAHAN MATRIKS DENGAN MENGGUNAKAN ATAU MEMANGGIL LIBRARY NUMPY:



```python
import numpy as np

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])

print('A+B=\n',a+b)
print('B+A=\n',b+a)
print('Sifat di atas merupakan sifat komutatif karena A+B = B+A')
```

BERIKUT CODE PERKALIAN MATRIKS DENGAN MENGGUNAKAN ATAU MEMANGGIL LIBRARY NUMPY:

##3.Perkalian Matriks:

Operasi ini melibatkan perkalian elemen-elemen matriks. Untuk melakukan perkalian matriks A dan B, jumlah kolom dalam matriks A harus sama dengan jumlah baris dalam matriks B. Contoh matriks A memiliki ordo atau ukuran mxn, maka matriks B harus memiliki ordo atau ukuran nxp. Sehingga hasil dari perkalian tersebut akan menghasilkan matriks baru dengan ukuran mxp. Hasilnya akan memiliki jumlah baris yang sama dengan matriks A dan jumlah kolom yang sama dengan matriks B.

CONTOH PERKALIAN MATRIKS:


\begin{align*}
A=
\begin{bmatrix}
2&3&2\\
4&1&-3\\
-1&-2&2
\end{bmatrix}
&&
B=
\begin{bmatrix}
2&4&7\\
6&-3&2\\
5&-2&7
\end{bmatrix}
\end{align*}

\begin{align*}
A X B =
\begin{bmatrix}
2(2)+3(6)+2(5) &2(4)+3(-3)+2(-2)&2(7)+3(2)+2(7)\\
4(2)+1(6)+(-3)(5) &4(4)+1(-3)+(-3)(-2) &4(7)+1(2)+(-3)(7)\\
-1(2)+(-2)(6)+2(5) &-1(4)+(-2)(-3)+2(-2) &-1(7)+(-2)(2)+2(7)
\end{bmatrix}
\end{align*}

\begin{align*}
AB=
\begin{bmatrix}
32&-5&34\\
-1&19&9\\
-4&-2&3
\end{bmatrix}
\end{align*}

BERIKUT CODE PERKALIAN MATRIKS DENGAN MENGGUNAKAN ATAU MEMANGGIL LIBRARY NUMPY:

```python
import numpy as np

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])
c=np.array([[3,4],[-2,4],[2,5]])

print('AB=\n',a@b)
print('AC=\n',a@c)
print('BC=\n',b@c)
```

- Sifat Asosiasi:

Dengan sifat asosiasi, urutan dalam melakukan operasi perhitungan tidak mempengaruhi hasil akhir.

  (AB)C = A(BC)

```python
import numpy as np

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])
c=np.array([[3,4],[-2,4],[2,5]])

print('A(BC) =\n', a@(b@c))
print('(AB)C =\n',(a@b)@c)
```

- Perkalian skalar:

Perkalian skalar adalah operasi yang melibatkan pengalihan suatu vektor oleh bilangan, yang disebut skalar. Biasanya skalar merupakan bilangan riil.

  c(AB) = (cA)B = A(cB)

```python
import numpy as np

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])
c=np.array([[3,4],[-2,4],[2,5]])

c = 3
print('c(AB)=\n',c*(a@b))
print('(cA)B=\n',(c*a)@b)
print('A(cB)=\n',a@(c*b))
```

- Perkalian identitas:

Dalam operasi perkalian ini, matriks dikalikan dengan matriks identitas. Dimana matriks identitas merupakan matriks persegi yang memiliki semua elemen di luar diagonal utama bernilai 1, dan elemen di luar diagonal utama bernilai 0. Matriks identitas dilambangkan dengan I. Contoh :

\begin{align*}
I=
\begin{bmatrix}
1&0&0\\
0&1&0\\
0&0&1
\end{bmatrix}
\end{align*}


  AI = IA = A

```python
import numpy as np

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])

i = np.array([[1,0,0],[0,1,0],[0,0,1]])
print('AI=\n',a@i)
print('IA=\n',i@a)
print('A=\n',a)
```

- Tidak berlaku sifat kumutatif:

Karena sifat komutatif membutuhkan bahwa operasi tidak dipengaruhi oleh urutan perkalian. Sementara matriks dipengaruhi oleh urutan perkalian. Sehingga:

  AB != BA atau AB tidak sama dengan BA

```python
import numpy as np

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])

print('AB=\n',a@b)
print('BA=\n',b@a)
```

##4.Transpose Matriks

Transpose matriks adalah matriks baru yang diperoleh dengan cara menukar elemen-elemen baris menjadi elemen kolom atau sebaliknya. Jika sebuah matriks berukuran M X N, maka transpose matriks tersebut memiliki ordo N X M.

CONTOH TRANSPOSE MATRIKS:

\begin{align*}
C=
\begin{bmatrix}
3&4\\
-2&4\\
2&5
\end{bmatrix}
&&
CT (C Transposre)=
\begin{bmatrix}
3&-2&2\\
4&4&-5
\end{bmatrix}
\end{align*}

BERIKUT CODE UNTUK MENG-TRANSPOSE MATRIKS DENGAN MENGGUNAKAN ATAU MEMANGGIL LIBRARY NUMPY:

```python
import numpy as np

c=np.array([[3,4],[-2,4],[2,5]])

c_tran = c.transpose()
print('matrik c sebelum di transpose=\n', c)
print('matrik c setelah di transpose=\n',c_tran)
print('ordo C_transpose adalah',c_tran.shape)
z=c_tran@c
print('Z = C_transpose x C =\n',z)
print('ordo z adalah',z.shape)
p=c@c_tran
print('P = C_transpose x C =\n',p)
print('ordo P adalah',p.shape)
print("Dapat disimpulkan bahwa C_transpose X C tidak sama dengan C X C_transpose")
```

Penjelasan :

- c_tran.shape

    - shape disini berfungsi untuk mengetahui ordo dari matriks c_tran dengan memanggil fungsi di dalam library numpy

#Program untuk Mengalikan Matriks dengan Menggunakan Perulangan (Loop) For :

```python
def perkalian(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        print("Perkalian matriks tidak dapat dilakukan karena jumlah kolom dalam matriks pertama tidak sama dengan jumlah baris dalam matriks kedua")
        return None
    result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    return result

def tampilan_hasil(matrix):
    for baris in matrix:
        print(baris)

print ("Berikut adalah contoh matriks")
print ("""
matriks 1=([[2,3,2],[4,1,-3],[-1,-2,2]])
matriks 1=([[2,4,7],[6,-3,2],[5,-2,7]])
""")
matriks1=([[2,3,2],[4,1,-3],[-1,-2,2]])
matriks2=([[2,4,7],[6,-3,2],[5,-2,7]])
test = perkalian(matriks1, matriks2)
print("\nHasil perkalian matriks:")
tampilan_hasil(test)
```
