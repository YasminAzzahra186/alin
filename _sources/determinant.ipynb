# DETERMINANT

Nilai determinan matriks memberikan informasi tentang apakah matriks tersebut memiliki invers (jika determinannya tidak nol), apakah matriks tersebut singuler atau tidak. Jika determinan suatu matriks adalah nol, maka matriks tersebut disebut sebagai matriks singular, yang artinya tidak memiliki invers. Sedangkan jika determinannya bukan nol, maka matriks tersebut disebut sebagai matriks nonsingular atau invertible, yang artinya memiliki invers.



## Syarat Mencari Determinan Matriks :
Hanya dapat dilakukan pada matriks bujur sangkar atau matriks persegi atau matriks yang mempunyai baris = kolom





## Sifat-sifat Determinan

1. Determinan dari matriks identitas = 1. Atau dapat ditulis det(I)=1
2. jika suatu matriks dikalikan dengan konstanta maka rumus dekterminannya:

    -> constanta^berapa baris atau kolom yang dikalikan dengan konstanta(hasil determinan dari matriks asalnya)

    atau
    
    -> K^n(det(matriks asal))

      keterangan :

      - K = konstanta

      - n = baris atau kolom yang telah dikalikan dengan dengan konstanta

      - det(matriks asal) = determinan dari matriks asal

3. jika matriks segitiga atas atau segitiga bawah dicari determinannya maka akan menghasilkan perkalian dari elemen di diagonalnya
4. $det(AB) = det(A)det(B)$
5. $det(A^-1) =  1/det(A)$
6. $det(A)^2 = det(A).det(A)$
7. $det(A^T) = det(A)$
  - Zero row -> zero column
  - Same row -> same column
8. $det(A+B) != det(A) + det(B)$

```python
import numpy as np

#fungsi mencari determinant
def det(matriks):
    return np.linalg.det(matriks)

print("Membuktikan bahwa determinan dari matriks identitas adalah 1")
I=np.array([[1,0,0],[0,1,0],[0,0,1]])
print('Matriks A =\n', I)
print("determinan dari matriks i adalah",det(I))

print("perubahan baris pada matriks identitas untuk mencari determinannya")
I2=np.array([[1,0,0],[0,0,1],[0,1,0]])
print("determinan dari matriks i setelah diubah pertama kali adalah",det(I2))
I3=np.array([[0,1,0],[0,0,1],[1,0,0]])
print("determinan dari matriks i setelah diubah kedua kali adalah",det(I3),"\n")

print("mencari determinan c")
c = np.array([[1,2,1],[3,1,2],[2,1,2]])
print("determinan dari matriks c adalah",det(c).round())
d = np.array([[3,1,2],[1,2,1],[2,1,2]])
print("determinan dari matriks c setelah diubah pertama kali adalah",det(d).round())
e = np.array([[1,2,1],[2,4,2],[2,1,2]])
print("determinan dari matriks e adalah",det(e).round())
print("jika baris dikalikan dengan kelipatan maka hasilnya akan 0")
f = np.array([[1,2,1],[2,4,2],[2,4,3]])
print("determinan dari matriks f adalah",det(f).round())
g = np.array([[6,2,4],[2,4,2],[4,2,4]])
print("determinan dari matriks g adalah",det(g).round())#2^3(3)-> 2=konstanta, ^3=jumlah baris yang dikalikan dengan konstanta, (3)=hasil dari det(c)
a=np.array([[3,4],[2,1]])
print("determinan dari matriks a adalah",det(a).round())
h=np.array([[6,8],[4,2]])
print("determinan dari matriks h adalah",det(h).round()) #2^2(-5)=-20
i=np.array([[6,8],[2,1]])
print("determinan dari matriks i adalah",det(i).round()) #jika hanya satu baris saja c det(a) = 2(-5) = -10

print("\nmengalikan salah satu kolom ")
j=np.array([[6,4],[4,1]])
print("determinan dari matriks j adalah",det(j).round())#hasilnya bakal sama dengan yang salah satu baris saja yang dikalikan
```

# COFACTOR

Adalah sebuah angka yang diperoleh dari elemen-elemen dalam matriks dengan cara tertenttu, yang sering digunakan dalam perhitungan determinan dan invers matriks. Cofactor dari $C_{ij}$ dari elemen $a_{ij}$ dalam sebuah matriks adalah hasil kali dari determinan submatriks yang diperoleh dengan menghapus baris $i$ dan kolom $j$ dari mattriks asli, dikalikan dengan $(-1)^{i+j}$


## Menghitung Invers Matriks 5 X 5 dengan cofactor

Kita bisa mencari invers matriks dengan menemukan determinan dan kofaktornya. Berikut langkah-langkah menghitung kofaktor:

1. Pilih elemen aij
2. Hapus baris dan kolom, baris ke-$i$ dan kolom ke-$j$ dari matriks untuk mendapatkan submatriks
3. Bentuk sub matriks, submatriks ini didapat dari elemen-elemen yang tersisa setelah menghapus baris dan kolom tersebut
4. Hitung determinan submatriks
5. Kalikan dengan $(-1)^{i+j}$, cofaktor $C_{ij}$ adalah hasil kali determinan submatriks dengan $(-1)^{i+j}$

Secara matematis, kofaktor Cij dari elemen aij adalah

$C_{ij} = (-1)^{i+j} .det(M_{ij})$

dimana det $(M_{ij})$ adalah determinan submatriks yang diperoleh dengan menghapus baris $i$ dan kolom $j$ dari matriks asli.

Rumus det pick row $i:$

$$det A = a_{i1}c_{i1}+a_{i2}c_{i2}+...+a_{in}c_{in}$$

Rumus det pick column $j:$

$$det A = a_{1j}c_{1j}+a_{2j}c_{2j}+...+a_{nj}c_{nj}$$

Rumus invers matrix:

$$A^-1= (1/det A) * adj(A) $$

Adjoin:

$$adj(A)=(kof(A))^T$$

(adj(A)) adalah hasil dari tranpose dari setiap cofactor-cofactor matriks

Contoh matriks 5X5:

$$
\begin{equation*}
Y=
\begin{bmatrix}
2 & 1 & 3 & 1 & 2 \\
1 & 1 & 6 & 1 & 3 \\
2 & -2 & 5 & 4 & 1 \\
1 & 2 & 1 & 3 & 1 \\
3 & 2 & 3 & 2 & 1 \\
\end{bmatrix}
\end{equation*}
$$

Berikut penyelesaian matriks di atas dengan memanggil library numpy:

```python
import numpy as np

#fungsi mencari determinan
def det(matriks):
    return np.linalg.det(matriks)
#fungsi untuk menghitung kofaktor
def hitung_kofaktor(matrix):
    def cofactor(matrix, i, j):
        minor = np.delete(np.delete(matrix, i, axis=0), j, axis=1)
        cof = (-1) ** (i + j) * np.linalg.det(minor)
        return cof
    rows, cols = matrix.shape
    cofactor_matrix = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            cofactor_matrix[i, j] = cofactor(matrix, i, j)
    return cofactor_matrix
#fungsi mencari invers
def invers (matriks):
  return np.linalg.inv(matriks)

Y = np.array ([[2,1,3,1,2],[1,1,6,1,3],[2,-2,5,4,1],[1,2,1,3,1],[3,2,3,2,1]])

#mencari sub matriks yang didapat dari elemen-elemen yang tersisa setelah menghapus baris dan kolom
#baris 1
Y11 = np.array([[1,6,1,3],[-2,5,4,1],[2,1,3,1],[2,3,2,1]])
Y12 = np.array([[1,6,1,3],[2,5,4,1],[1,1,3,1],[3,3,2,1]])
Y13 = np.array([[1,1,1,3],[2,-2,4,1],[1,2,3,1],[3,2,2,1]])
Y14 = np.array([[1,1,6,3],[2,-2,5,1],[1,2,1,1],[3,2,3,1]])
Y15 = np.array([[1,1,6,1],[2,-2,5,4],[1,2,1,3],[3,2,3,2]])
#baris 2
Y21 = np.array([[1,3,1,2],[-2,5,4,1],[2,1,3,1],[2,3,2,1]])
Y22 = np.array([[2,3,1,2],[2,5,4,1],[1,1,3,1],[3,3,2,1]])
Y23 = np.array([[2,1,1,2],[2,-2,4,1],[1,2,3,1],[3,2,2,1]])
Y24 = np.array([[2,1,3,2],[2,-2,5,1],[1,2,1,1],[3,2,3,1]])
Y25 = np.array([[2,1,3,1],[2,-2,5,4],[1,2,1,3],[3,2,3,2]])
#baris 3
Y31 = np.array([[1,3,1,2],[1,6,1,3],[2,1,3,1],[2,3,2,1]])
Y32 = np.array([[2,3,1,2],[1,6,1,3],[1,1,3,1],[3,3,2,1]])
Y33 = np.array([[2,1,1,2],[1,1,1,3],[1,2,3,1],[3,2,2,1]])
Y34 = np.array([[2,1,3,2],[1,1,6,3],[1,2,1,1],[3,2,3,1]])
Y35 = np.array([[2,1,3,1],[1,1,6,1],[1,2,1,3],[3,2,3,2]])
#baris 4
Y41 = np.array([[1,3,1,2],[1,6,1,3],[-2,5,4,1],[2,3,2,1]])
Y42 = np.array([[2,3,1,2],[1,6,1,3],[2,5,4,1],[3,3,2,1]])
Y43 = np.array([[2,1,1,2],[1,1,1,3],[2,-2,4,1],[3,2,2,1]])
Y44 = np.array([[2,1,3,2],[1,1,6,3],[2,-2,5,1],[3,2,3,1]])
Y45 = np.array([[2,1,3,1],[1,1,6,1],[2,-2,5,4,],[3,2,3,2]])
#baris 5
Y51 = np.array([[1,3,1,2],[1,6,1,3],[-2,5,4,1],[2,1,3,1]])
Y52 = np.array([[2,3,1,2],[1,6,1,3],[2,5,4,1],[1,1,3,1]])
Y53 = np.array([[2,1,1,2],[1,1,1,3],[2,-2,4,1],[1,2,3,1]])
Y54 = np.array([[2,1,3,2],[1,1,6,3],[2,-2,5,1],[1,2,1,1]])
Y55 = np.array([[2,1,3,1],[1,1,6,1],[2,-2,5,4],[1,2,1,3]])
```
