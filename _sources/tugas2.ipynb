#Operasi pada Matriks
contoh matriks sebagai berikut:

```python
import numpy as np
import sympy as sy

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])
c=np.array([[3,4],[-2,4],[2,5]])
print('A=\n',a)
print('B=\n',b)
print('C=\n',c)
```

##1.Penjumlahan matriks :



  Note!!
  
  penjumlahan dan pengurangan matriks yaitu matriks yang akan dijumlahkan atau dikurangkan harus memiliki ordo atau ukuran yang sama. Artinya, jumlah baris dan kolomnya harus sama.

```python
print('A+B=\n',a+b)
print('B+A=\n',b+a)
print('Sifat di atas merupakan sifat komutatif karena A+B = B+A')
```

##2.Pengurangan Matriks:

```python
print('A-B=\n',a-b)
print('B-A=\n',b-a)
print('Sifat di atas merupakan sifat komutatif karena A-B = B-A')
```

##3.Perkalian Matriks:

Note!!

Untuk melakukan perkalian matriks A dan B, jumlah kolom dalam matriks A harus sama dengan jumlah baris dalam matriks B. Contoh matriks A memiliki ordo atau ukuran mxn, maka matriks B harus memiliki ordo atau ukuran nxp. Sehingga hasil dari perkalian tersebut akan menghasilkan matriks baru dengan ukuran mxp. Hasilnya akan memiliki jumlah baris yang sama dengan matriks A dan jumlah kolom yang sama dengan matriks B.

```python
print('AB=\n',a@b)
print('AC=\n',a@c)
print('BC=\n',b@c)
```

- Sifat Asosiasi:

  (AB)C = A(BC)

```python
print('A(BC)\n',a@(b@c))
print('(AB)C =\n',(a@b)@c)
```

- Perkalian skalar:

  c(AB) = (cA)B = A(cB)

```python
c = 3
print('c(AB)\n',c*(a@b))
print('(cA)B\n',(c*a)@b)
print('A(cB)\n',a@(c*b))
```

- Perkalian identitas:

  AI = IA = A

```python
i = np.array([[1,0,0],[0,1,0],[0,0,1]])
print('AI=\n',a@i)
print('IA=\n',i@a)
print('A=\n',a)
```

- Perkalian nol:

  A0 = 0A = A

```python
print('A0=\n',a*0)
print('0A=\n',0*a)
print('A=\n',a)
```

- Tidak berlaku sifat kumutatif:

  AB != BA atau AB tidak sama dengan BA

```python
print('AB=\n',a@b)
print('BA=\n',b@a)
```

##4.Transpose Matriks

```python
c_tran = c.transpose()
print('matrik c sebelum di transpose=\n', c)
print('matrik c setelah di transpose=\n',c_tran)
print('ordo C_transpose adalah',c_tran.shape)
z=c_tran@c
print('Z = C_transpose x C =\n',z)
print('ordo z adalah',z.shape)
p=c@c_tran
print('P = C_transpose x C =\n',p)
print('ordo P adalah',p.shape)
```
