#Penjelasan Mengenai Matriks
##Pengertian Matriks:
Matriks adalah susunan bilagan real(atau elemen-elemen) yang disusun dalam baris dan kolom sehingga membentuk persegi ataupun persegi panjang. Setiap elemen memiliki posisi unik dalam matriks, ditentukan oleh baris dan kolomnya. Misalnya, matriks 2x3 memiliki 2 baris dan 3 kolom. suatu matriks biasa diberi nama dengan menggunakan huruf kapital (seperti A, B dll) dan elemennya dinyatakan dengan huruf kecil.


  matriks.png
##Jenis - jenis Matriks:
1. Matriks nol, matriks yang semua elemen
penyusunnnya nol.
2. Matriks baris, matriks yang elemen penyusunnya terdiri dari satu baris.
3. Matriks kolom, matriks yang elemen penyusunnya terdiri, satu kolom.
4. Matriks persegi adalah matriks yang mempunyai banyak baris sama dengan banyaknya kolom.
5. Matriks diagonal, matriks yang mempunyai elemen selain diagonal utamanya.
6. Matriks skalar, matriks yang elemen pada diagonal utamanya sama dengan satu.
7. Matriks identitas, matriks skalar yang elemenpada diagonal utamanya sama dengan satu.
8. Matriks simatri, matriks persegi yang setiap elemennya selain elemen diagonal adalah simetri terhadap diagonal utama.
9. Matriks simateri miring, matriks yang elemen-elemennya, selain elemen diagonal salinh berlawanan.
10. Matriks segitiga atas, matriks persegi yang elemen-elemen di bawah diagonal utamanya adalah nol.
11. Matriks segitiga bawah, matriks persegi yang elem-elemen di atas diagonal utamanya adalah nol.
12. Matriks tranpose, matriksyang diperoleh dari memindahkan elemen-elemen baris menjadi elemen pada kolom atau sebaliknya.



#Operasi pada Matriks
contoh matriks sebagai berikut:

```python
import numpy as np
import sympy as sy

a=np.array([[2,3,2],[4,1,-3],[-1,-2,2]])
b=np.array([[2,4,7],[6,-3,2],[5,-2,7]])
c=np.array([[3,4],[-2,4],[2,5]])
print('A=\n',a)
print('B=\n',b)
print('C=\n',c)
```

##1.Penjumlahan matriks :

  Note!!
   
   Penjumlahan dan pengurangan matriks yaitu matriks yang akan dijumlahkan atau dikurangkan harus memiliki ordo atau ukuran yang sama. Artinya, jumlah baris dan kolomnya harus sama.



```python
print('A+B=\n',a+b)
print('B+A=\n',b+a)
print('Sifat di atas merupakan sifat komutatif karena A+B = B+A')
```

##2.Pengurangan Matriks:

```python
print('A-B=\n',a-b)
print('B-A=\n',b-a)
print('Sifat di atas merupakan sifat komutatif karena A-B = B-A')
```

##3.Perkalian Matriks:

Note!!

Untuk melakukan perkalian matriks A dan B, jumlah kolom dalam matriks A harus sama dengan jumlah baris dalam matriks B. Contoh matriks A memiliki ordo atau ukuran mxn, maka matriks B harus memiliki ordo atau ukuran nxp. Sehingga hasil dari perkalian tersebut akan menghasilkan matriks baru dengan ukuran mxp. Hasilnya akan memiliki jumlah baris yang sama dengan matriks A dan jumlah kolom yang sama dengan matriks B.

```python
print('AB=\n',a@b)
print('AC=\n',a@c)
print('BC=\n',b@c)
```

- Sifat Asosiasi:

  (AB)C = A(BC)

```python
print('A(BC) =\n', a@(b@c))
print('(AB)C =\n',(a@b)@c)
```

- Perkalian skalar:

  c(AB) = (cA)B = A(cB)

```python
c = 3
print('c(AB)\n',c*(a@b))
print('(cA)B\n',(c*a)@b)
print('A(cB)\n',a@(c*b))
```

- Perkalian identitas:

  AI = IA = A

```python
i = np.array([[1,0,0],[0,1,0],[0,0,1]])
print('AI=\n',a@i)
print('IA=\n',i@a)
print('A=\n',a)
```

- Tidak berlaku sifat kumutatif:

  AB != BA atau AB tidak sama dengan BA

```python
print('AB=\n',a@b)
print('BA=\n',b@a)
```

##4.Transpose Matriks
Adalah Transpose matriks adalah matriks baru yang diperoleh dengan cara menukar elemen-elemen baris menjadi elemen kolom atau sebaliknya. Jika sebuah matriks berukuran M X N, maka transpose matriks tersebut memiliki ordo N X M.

Berikut contohnya:

```python
c=np.array([[3,4],[-2,4],[2,5]])
c_tran = c.transpose()
print('matrik c sebelum di transpose=\n', c)
print('matrik c setelah di transpose=\n',c_tran)
print('ordo C_transpose adalah',c_tran.shape)
z=c_tran@c
print('Z = C_transpose x C =\n',z)
print('ordo z adalah',z.shape)
p=c@c_tran
print('P = C_transpose x C =\n',p)
print('ordo P adalah',p.shape)
print("Dapat disimpulkan bahwa C_transpose X C tidak sama dengan C X C_transpose")
```

#Program untuk Mengalikan Matriks

```python
def input_matrix():
    baris = int(input("Masukkan jumlah baris: "))
    kolom = int(input("Masukkan jumlah kolom: "))
    matrix = []
    for i in range(baris):
        row = []
        print(f"Masukkan elemen untuk baris {i + 1}: ")
        for j in range(kolom):
            element = float(input(f"Elemen [{i + 1}, {j + 1}]: "))
            row.append(element)
        matrix.append(row)
    return matrix

def perkalian(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        print("Perkalian matriks tidak dapat dilakukan karena jumlah kolom dalam matriks pertama tidak sama dengan jumlah baris dalam matriks kedua")
        return None
    result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    return result

def tampilan_hasil(matrix):
    for baris in matrix:
        print(baris)
coba = input("Ingin mengisi matriks sendiri? (iya/tidak) : ")
if coba == "iya":
    print("Program perkalian matriks!!")
    print("Masukkan matriks pertama:")
    matriks1 = input_matrix()
    print ("Matriks 1 = ", matriks1)
    print("Masukkan matriks kedua:")
    matriks2 = input_matrix()
    print ("Matriks 2 = ",matriks2)
elif coba == "tidak":
    print ("Berikut adalah contoh matriks")
    print ("""
    matriks 1=([[2,3,2],[4,1,-3],[-1,-2,2]])
    matriks 1=([[2,4,7],[6,-3,2],[5,-2,7]])
    """)
    matriks1=([[2,3,2],[4,1,-3],[-1,-2,2]])
    matriks2=([[2,4,7],[6,-3,2],[5,-2,7]])
else:
    print("Maaf pilihan ersebut tidak tersedia. Tolong masukkan pilihan yang sesuai")

test = perkalian(matriks1, matriks2)
print("\nHasil perkalian matriks:")
tampilan_hasil(test)
```
