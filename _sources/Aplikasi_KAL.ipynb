# Page Rank

PageRank adalah sebuah algoritma yang dikembangkan oleh Larry Page dan Sergey Brin, yang merupakan pendiri Google, pada akhir 1990-an. Algoritma ini digunakan untuk menentukan peringkat halaman web dalam hasil pencarian mesin pencari. Page Rank bekerja dengan cara menghitung jumlah dan kualitas dari link yang mengarah ke web page tersebut untuk menentukan perkiraan seberapa penting web pages tersebut. Bisa diasumsikan dengan semakin penting web pages tersebut maka semakin banyak link yang menuju ke web pages tersebut. Inti dari PageRank adalah untuk mengukur pentingnya sebuah halaman web berdasarkan jumlah dan kualitas tautan (link) yang mengarah ke halaman tersebut.

Page Rank dihasilkan dari Algoritma Matematika yaitu web graph yang dipakai seluruh world wide web pages sebagai node, dan hyperlink sebagai edge.

## Algoritma Pagerank

Penjelasan singkat mengenai bagaimana PageRank bekerja:

a. Prinsip Dasar: PageRank bekerja dengan asumsi bahwa tautan dari satu halaman web ke halaman web lainnya dapat dianggap sebagai “suara” untuk halaman tujuan. Semakin banyak suara (tautan) yang diterima oleh sebuah halaman, semakin tinggi peringkatnya.

b. Bobot Tautan: Tidak semua tautan dianggap sama. Tautan dari halaman yang memiliki PageRank tinggi memberikan kontribusi lebih besar dibandingkan tautan dari halaman dengan PageRank rendah. Jadi, tautan dari situs yang sudah terkenal atau sering dikunjungi akan memberikan pengaruh lebih besar.

c. Distribusi PageRank: Setiap halaman web diberikan nilai awal yang sama. PageRank dari sebuah halaman akan dibagi secara merata ke semua halaman yang ditautkan dari halaman tersebut. Kemudian, PageRank dari semua halaman akan dihitung ulang secara iteratif hingga mencapai keseimbangan (konvergen).

d. Damping Factor: Algoritma PageRank menggunakan faktor redaman (damping factor) untuk menambahkan elemen probabilistik. Faktor ini biasanya diatur sekitar 0.85, yang berarti ada 85% kemungkinan bahwa pengguna akan mengikuti tautan di halaman tersebut, dan 15% kemungkinan mereka akan langsung menuju halaman lain secara acak. Ini membantu mencegah beberapa halaman menjadi terlalu dominan dan memastikan bahwa setiap halaman memiliki kesempatan untuk mendapatkan peringkat.

PageRank menjadi salah satu faktor utama dalam algoritma pencarian Google karena mampu memberikan hasil yang relevan dan berkualitas tinggi dengan memperhitungkan popularitas dan otoritas halaman web berdasarkan jaringan tautan. Namun, seiring waktu, Google telah menambahkan banyak faktor lain dalam algoritma pencariannya untuk meningkatkan hasil pencarian, meskipun konsep dasar PageRank masih tetap relevan.

Algoritma Pagerank dimulai dengan menetapkan beberapa nilai awal Page Rank kepada seluruh Web Pages. Algoritma tersebut kemudian diterapkan secara iteratif hingga mencapai kondisi stabil/tetap/memberikan sedikit/tidak ada perubahan lebih lanjut dalam distribusi PageRank. Nilai awal Page Rank merupakan fungsi dari jumlah halaman dalam indeks. Untuk Nilai Page Rank Asli adalah $1/N$ untuk $N$ jumlah halaman dalam indeks.

## Berikut Contoh Penerapan untuk Mencari Pagerank dalam Code

Menggunakan cara dengan mengalikan secara satu persatu sehingga hasilnya tidak jauh berbeda dengan hasil sebelumnya

```python

import numpy as np

A = np.array ([[0,0,1,1/2],[1/3,0,0,0],[1/3,1/2,0,1/2],[1/3,1/2,0,0]])

v = np.array ([[0.25],[0.25],[0.25],[0.25]])

print("Av = \n",A@v)
print("\n(A^2)v = \n", A@A@v)
print("\n(A^3)v = \n", A@A@A@v)
print("\n(A^4)v = \n", A@A@A@A@v)
print("\n(A^5)v = \n", A@A@A@A@A@v)
print("\n(A^6)v = \n", A@A@A@A@A@A@v)
print("\n(A^7)v = \n", A@A@A@A@A@A@A@v)
print("\n(A^8)v = \n", A@A@A@A@A@A@A@A@v)
```

## Perkalian matriks konvergen

```python
import numpy as np

A = np.array([[0, 0, 1, 1/2],
              [1/3, 0, 0, 0],
              [1/3, 1/2, 0, 1/2],
              [1/3, 1/2, 0, 0]])

v = np.array([[0.20], [0.30], [0.25], [0.25]])

previous_result = v
tolerance = 0.01

for i in range(1, 100):
    result = A @ previous_result
    print(f"(A^{i})v = \n", result)
    if np.all(np.abs(result - previous_result) <= tolerance):
        if np.sqrt(np.sum((result - previous_result) ** 2)) < tolerance:
          print(f"\nPerkalian matriks konvergen pada iterasi ke-{i}.")
          break

    previous_result = result
```

Keterangan Code di Atas:

- 'import numpy as np'

    -> Mengimpor pustaka numpy yang digunakan untuk operasi numerik dalam Python.

- 'A = np.array([[0, 0, 1, 1/2], [1/3, 0, 0, 0], [1/3, 1/2, 0, 1/2], [1/3, 1/2, 0, 0]])'

    -> Mendefinisikan matriks A sebagai array numpy 4x4. Setiap elemen dalam matriks ini merepresentasikan bobot transisi dari satu keadaan ke keadaan lainnya.

- 'v = np.array([[0.40], [0.30], [0.20], [0.10]])'

    -> Mendefinisikan vektor v sebagai array numpy 4x1 dengan nilai-nilai yang berbeda. Ini adalah vektor awal yang akan digunakan dalam iterasi. Setiap elemen dalam vektor ini bisa merepresentasikan distribusi probabilitas awal dari suatu proses.

- 'previous_result = v'

    -> previous_result diinisialisasi dengan nilai v. Ini akan menyimpan hasil dari iterasi sebelumnya.

- 'tolerance = 0.01'

    -> tolerance ditetapkan menjadi 0.01, yang menentukan batas perbedaan antara hasil iterasi berturut-turut untuk menentukan konvergensi. Semakin kecil nilai tolerance maka semakin mendekati valid.

- 'for i in range(1, 100):'

    -> Memulai loop for yang akan berjalan dari 1 hingga 99 (total 99 iterasi) untuk mencegah loop tak terbatas. Loop ini digunakan untuk memperbarui hasil perkalian matriks dan memeriksa konvergensi.

- 'result = A @ previous_result'

    -> result dihitung dengan mengalikan matriks A dengan previous_result menggunakan operator @ (matrix multiplication).

- 'print(f"(A^{i})v = \n", result)'

    -> Hasil perkalian matriks ditampilkan di layar dengan memberikan informasi iterasi ke-i.

- 'np.abs(result - previous_result)'

    -> menghitung perbedaan absolut antara result dan previous_result.

- 'np.all(np.abs(result - previous_result) <= tolerance)'

    -> memeriksa apakah semua elemen dalam hasil perbedaan kurang dari atau sama dengan tolerance.

- 'np.sqrt(np.sum((result - previous_result) ** 2)) < tolerance'

    -> menghitung norma Euclidean dari perbedaan dan memeriksa apakah kurang dari tolerance.

- 'print(f"\nPerkalian matriks konvergen pada iterasi ke-{i}.")'

    -> pesan bahwa matriks telah konvergen pada iterasi ke-i.

- 'break'

    -> iterasi dihentikan

- 'previous_result = result'

    -> previous_result diperbarui dengan result untuk digunakan dalam iterasi berikutnya.
