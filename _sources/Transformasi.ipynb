# Transformasi Linier

## Definisi Transformasi

Fungsi f dari himpunan X ke himpunan Y terdapat aturan yang menetapkan setiap elemen Y ke setiap elemen X. Misalkan himpunan X dan Y keduanya merupakan himpunan semua vektor dalam ruang, dapat menyatakan sebagai suatu transformasi.

    Transformasi adalah pemetaan T dari ruang vektor V ke ruang vektor W .

    v ∈ −−−−−Transformation−−−−→ w = T (v) ∈ W

```python
import numpy as np

# Definisikan matriks A
A = np.array([[2, 0, 0],
              [0, 1, 0],
              [0, 0, 1]])

# Fungsi transformasi linier
def linear_transformation(x):
    return np.dot(A, x)

# Contoh vektor input
x = np.array([2, 3, 1])

# Hitung hasil transformasi
result = linear_transformation(x)

print("Hasil transformasi:", result)
```

## Definisi Transformasi Linier pada matriks

Transformasi linear pada matriks mengacu pada penggunaan matriks untuk merepresentasikan transformasi linear dari satu ruang ke ruang lain. Transformasi linear adalah fungsi matematika yang mempertahankan sifat struktural ruang vektor, yaitu penjumlahan dan perkalian skalar.

## Sifat-sifat Transformasi Linier:
- Memelihara asal: Transformasi linier memetakan vektor nol ke vektor nol.
- Memelihara kombinasi linear: Transformasi linier memetakan kombinasi linear dari vektor-vektor ke kombinasi linear dari bayangan vektor-vektor tersebut.
- Representasi matriks: cara untuk menggambarkan atau menyajikan sebuah matriks dalam bentuk yang dapat dimengerti. Contohnya :
    - Notasi Matematika, notasi matematika standar yang menggunakan tanda kurung dan indeks untuk menyatakan elemen-elemennya.
    - Array, struktur data dua dimensi. Setiap elemen dalam array menyimpan nilai dari elemen matriks yang sesuai. Representasi ini sangat umum digunakan dalam komputasi numerik dan ilmu komputer.
    - Visualisasi Grafis, Untuk matriks dengan dimensi yang kecil, terkadang berguna untuk memvisualisasikannya dalam bentuk tabel atau grafik. Ini dapat membantu dalam pemahaman intuitif tentang struktur matriks dan pola data yang terkandung di dalamnya.
    - Matriks Blok, matriks dipecah-pecah. Representasi ini berguna dalam beberapa kasus di mana matriks besar dapat diurai menjadi bagian-bagian yang lebih mudah dikelola.
    - Sparse Matriks Reprentation, Untuk matriks yang memiliki banyak elemen nol, seringkali lebih efisien untuk hanya menyimpan elemen-elemen yang tidak nol dan posisi mereka dalam matriks.

### Contoh transformasi linier dalam matriks:
- Rotasi: Memutar vektor di sekitar sumbu tertentu. Matriks rotasi memiliki elemen yang ditentukan oleh sudut rotasi dan sumbu rotasi.

### Cara Menentukan Transformasi Linier

Dua aturan:
- Pusat titik tidak berubah.
- Garis lurus ditetapkan sebagai garis lurus.

Secara umum pemetaan T adalah transformasi linier jika :

### 1. T (u + v) = T (u) + T (v)

```python
import numpy as np

# Definisikan matriks transformasi T
T = np.array([[1, 2],
              [3, 4]])

# Fungsi untuk transformasi linier
def linear_transformation(T, u, v):
    # Menghitung T(u + v)
    Tu_plus_v = np.dot(T, u + v)

    # Menghitung T(u) + T(v)
    Tu_plus_Tv = np.dot(T, u) + np.dot(T, v)

    # Memeriksa apakah T(u + v) sama dengan T(u) + T(v)
    if np.array_equal(Tu_plus_v, Tu_plus_Tv):
        print("Transformasi linier memenuhi sifat T(u + v) = T(u) + T(v)")
    else:
        print("Transformasi linier tidak memenuhi sifat T(u + v) = T(u) + T(v)")

    return Tu_plus_v, Tu_plus_Tv

# Contoh vektor u dan v
u = np.array([2, 1])
v = np.array([1, 3])

# Panggil fungsi dan cetak hasilnya
Tu_plus_v, Tu_plus_Tv = linear_transformation(T, u, v)
print("T(u + v):", Tu_plus_v)
print("T(u) + T(v):", Tu_plus_Tv)
```

### 2. T (αv) = αT (v)

```python
import numpy as np

# Definisikan matriks transformasi T
T = np.array([[1, 2],
              [3, 4]])

# Fungsi untuk transformasi linier
def linear_transformation(T, v):
    return np.dot(T, v)

# Contoh vektor v dan skalar alpha
v = np.array([1, 3])
alpha = 2

# Hitung T(alpha * v)
T_alpha_v = linear_transformation(T, alpha * v)

# Hitung alpha * T(v)
alpha_T_v = alpha * linear_transformation(T, v)

# Memeriksa apakah T(alpha * v) sama dengan alpha * T(v)
if np.array_equal(T_alpha_v, alpha_T_v):
    print("Transformasi linier memenuhi sifat T(alpha * v) = alpha * T(v)")
else:
    print("Transformasi linier tidak memenuhi sifat T(alpha * v) = alpha * T(v)")

print("T(alpha * v):", T_alpha_v)
print("alpha * T(v):", alpha_T_v)
```

### Refleksi Matriks
Tranformasi Linier dengan Refleksi atau pencerminan pada tranformasi linear adalah adalah salah satu metode tranformasi linear yang digunakan untuk mencerminkan suatu Node dengan fungsi matrix. Berikut adalah sifat-sifat Refleksi:

1. Objek yang direfleksikan tidak mengalami perubahan benntuk dan ukuran

2. Jarak objek dari cermin (sumbu X/Y) adalah sama dengan jarak bayangan dengan cermin tersbut.

- a. Untuk bisa dicerminkan ke sumbu X fungsi matrix harus seperti berikut:

$$
\begin{equation*}
v=
\begin{bmatrix}
1&0\\0&-1
\end{bmatrix}
\end{equation*}
$$

- b. Untuk bisa dicerminkan ke sumbu Y fungsi matrix harus seperti berikut:

$$
\begin{equation*}
v'=
\begin{bmatrix}
-1&0\\0&1
\end{bmatrix}
\end{equation*}
$$


#### Refleksi terhadap sumbu x

```python
import numpy as np

# Membuat matriks 2x2 dengan nilai awal
matriks = np.array([[1, 2], [3, 4]])

# Melakukan cermin pada sumbu x dengan mengalikan baris kedua dengan -1
matriks[1, :] = matriks[1, :] * -1

# Menampilkan matriks setelah transformasi cermin
print("Matriks setelah transformasi cermin pada sumbu x:")
print(matriks)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks refleksi terhadap sumbu x
reflection_matrix = np.array([[1, 2], [-3, -4]])

# Vektor asli
original_vector = np.array([1, 3])

# Menerapkan refleksi menggunakan matriks refleksi
reflected_vector = np.dot(reflection_matrix, original_vector)

# Plot vektor asli dan hasil refleksi
plt.figure()
plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vektor Asli')
plt.quiver(0, 0, reflected_vector[0], reflected_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Hasil Refleksi (Sumbu X)')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Refleksi Matriks Identitas (Sumbu X)')
plt.legend()
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

```

#### Refleksi terhadap sumbu y

```python
import numpy as np

# Membuat matriks 2x2 dengan nilai awal
matriks = np.array([[1, 2], [3, 4]])

# Melakukan cermin pada sumbu y dengan mengalikan kolom pertama dengan -1
matriks[:, 0] = matriks[:, 0] * -1

# Menampilkan matriks setelah transformasi cermin
print("Matriks setelah transformasi cermin pada sumbu y:")
print(matriks)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks identitas
identity_matrix = np.array([[1, 0], [0, 1]])

# Vektor asli
original_vector = np.array([1, 3])

# Menerapkan proyeksi menggunakan matriks identitas
projected_vector = np.dot(identity_matrix, original_vector)

# Merefleksikan hasil proyeksi pada sumbu y
reflected_vector = np.array([-projected_vector[0], projected_vector[1]])

# Plot vektor asli dan hasil proyeksi
plt.figure()
plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vektor Asli')
plt.quiver(0, 0, reflected_vector[0], reflected_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Hasil Proyeksi')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Proyeksi Matriks Identitas')
plt.legend()
plt.grid()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Rotasi Matriks
Rotasi matriks adalah jenis transformasi linier yang digunakan untuk memutar objek dalam ruang sebesar sudut tertentu, baik dalam dimensi 2D maupun 3D. Transformasi rotasi sangat berguna dalam berbagai aplikasi seperti grafika komputer, animasi, pemrosesan citra, dan rekayasa. Rotasi matriks diaplikasikan pada vektor-vektor yang merepresentasikan posisi objek dalam ruang. Hasil rotasi dari matriks rotasi adalah vektor yang merepresentasikan posisi objek setelah rotasi diaplikasikan.

Rotasi matriks umumnya dinyatakan dalam dua dimensi dan tiga dimensi:
- Dalam dimensi 2:
$$
\begin{align*}
R=
\begin{bmatrix}
cos(θ)&-sin(θ)\\
sin(θ)&cos(θ)\\
\end{bmatrix}
\end{align*}
$$

Di sini, $θ$ adalah sudut rotasi yang diberikan dalam radian. Matriks rotasi ini akan memutar objek sebesar $θ$ searah jarum jam jika $θ$ positif, atau berlawanan arah jarum jam jika $θ$ negatif.

- Dalam dimensi 3:

Rotasi dalam tiga dimensi lebih kompleks karena dapat terjadi rotasi di sekitar sumbu yang berbeda. Misalnya, rotasi bisa terjadi di sekitar sumbu-x, sumbu-y, atau sumbu-z.



```python
import numpy as np

# Membuat matriks 2x2 dengan nilai awal
matriks = np.array([[1, 2], [3, 4]])

# Mendefinisikan matriks rotasi untuk rotasi 90 derajat searah jarum jam
theta = np.radians(90)  # Mengubah derajat menjadi radian
rotasi_matriks = np.array([[np.cos(theta), -np.sin(theta)],
                           [np.sin(theta), np.cos(theta)]])

# Melakukan rotasi matriks
matriks_rotasi = np.dot(rotasi_matriks, matriks)

# Menampilkan matriks setelah rotasi
print("Matriks setelah rotasi 90 derajat searah jarum jam:")
print(matriks_rotasi.round())
```

```python
import numpy as np

# Membuat matriks 2x2 dengan nilai awal
matriks = np.array([[1, 2], [3, 4]])

# Melakukan rotasi 90 derajat berlawanan arah jarum jam
matriks_rotasi = np.rot90(matriks)

# Menampilkan matriks setelah rotasi
print("Matriks setelah rotasi 90 derajat berlawanan arah jarum jam:")
print(matriks_rotasi)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sudut rotasi dalam radian (misalnya, 45 derajat)
theta = np.radians(45)

# Matriks rotasi
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

# Vektor asli
original_vector = np.array([1, 3])

# Menerapkan rotasi menggunakan matriks
rotated_vector = np.dot(rotation_matrix, original_vector)

# Plot vektor asli dan hasil rotasi
plt.figure()
plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vektor Asli')
plt.quiver(0, 0, rotated_vector[0], rotated_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Hasil Rotasi')
plt.xlim(-5, 5)  # Sesuaikan batas sumbu x sesuai kebutuhan
plt.ylim(-5, 5)  # Sesuaikan batas sumbu y sesuai kebutuhan
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Rotasi Matriks')
plt.legend()
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Translasi Matriks
Translasi matriks adalah jenis transformasi linier yang digunakan untuk menggeser objek dalam ruang. Transformasi translasi berguna untuk memindahkan objek dari satu lokasi ke lokasi lain dalam koordinat 2D atau 3D.

Matriks translasi umumnya memiliki bentuk seperti ini:
- Dalam dimensi 2:
$$
\begin{align*}
A=
\begin{bmatrix}
1&0&t_x\\
0&1&t_y\\
0&0&1
\end{bmatrix}
\end{align*}
$$

Di sini, $t_x$ dan $t_y$ adalah pergeseran yang diberikan dalam sumbu-x dan sumbu-y, secara berturut-turut.

- Dalam dimensi 3:
$$
\begin{align*}
A=
\begin{bmatrix}
1&0&0&t_x\\
0&1&0&t_y\\
0&0&1&t_z\\
0&0&0&1
\end{bmatrix}
\end{align*}
$$

Di sini, $t_x,t_y,$ dan $t_z$ adalah pergeseran yang diberikan dalam sumbu-x, sumbu-y, dan sumbu-z, secara berturut-turut.

Ketika matriks translasi ini diterapkan pada vektor $(x,y)$ atau $(x,y,z)$, koordinat objek akan ditambahkan dengan nilai $t_x,t_y,$ dan $t_z$ masing-masing, sehingga menghasilkan pergeseran sesuai dengan yang diinginkan.

Contoh memiliki vektor =

$$
\begin{equation*}
v=
\begin{bmatrix}
1\\3
\end{bmatrix}
\end{equation*}
$$

Translasi sebesar


$$
\begin{bmatrix}
2\\1
\end{bmatrix}
$$

akan menghasilkan vektor

$$
\begin{equation*}
v'=
\begin{bmatrix}
3\\4
\end{bmatrix}
\end{equation*}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks translasi
translation_matrix = np.array([[1, 0, 1], [0, 3, 1], [0, 0, 1]])  # Translasi sebesar (2, 3)

# Vektor asli (ditambahkan dimensi ke-3 untuk koordinat homogen)
original_vector_homogeneous = np.array([1, 3, 1])

# Menerapkan translasi menggunakan matriks
translated_vector_homogeneous = np.dot(translation_matrix, original_vector_homogeneous)

# Ambil hasil translasi dari koordinat homogen dan hilangkan dimensi ke-3
translated_vector = translated_vector_homogeneous[:2]

# Plot vektor asli dan hasil translasi
plt.figure()
plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vektor Asli')
plt.quiver(0, 0, translated_vector[0], translated_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Hasil Translasi')
plt.xlim(-5, 7)  # Sesuaikan batas sumbu x sesuai dengan translasi yang dilakukan
plt.ylim(-5, 7)  # Sesuaikan batas sumbu y sesuai dengan translasi yang dilakukan
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Translasi Matriks')
plt.legend()
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Dilatasi Matriks
Transformasi linier scaling adalah jenis transformasi linear di mana setiap koordinat dari objek digandakan atau dikurangi dengan faktor skalar tertentu. Ini dilakukan dengan memperbesar (scaling up) atau memperkecil (scaling down) objek dalam ruang. Misalnya, dalam 2 dimensi, scaling akan mengubah ukuran objek di sepanjang sumbu x dan sumbu y.

Transformasi linier scaling dalam matriks dapat direpresentasikan dengan matriks transformasi khusus. Misalkan $v_x$ dan $v_y$  adalah faktor skalar untuk sumbu x dan sumbu y, masing-masing. Maka, matriks transformasi linier scaling akan memiliki bentuk seperti ini:

$$
\begin{equation*}
v=
\begin{bmatrix}
v_x&0\\0&v_y
\end{bmatrix}
\end{equation*}
$$

Ketika matriks transformasi ini diterapkan pada vektor ($x, y$) hasilnya adalah:

$$
\begin{align*}
\begin{bmatrix}
v_x&0\\
0&v_y
\end{bmatrix}
\begin{bmatrix}
x\\
y\\
\end{bmatrix}
=
\begin{bmatrix}
v_x.x\\
v_y.y\\
\end{bmatrix}
\end{align*}
$$

Ini artinya, koordinat $(x,y)$ akan digandakan atau dikurangi dengan faktor skalar $v_x$ dan $v_y$, masing-masing, untuk mendapatkan koordinat baru $(v_x.x,v_y.y)$.

Contoh penggunaan scaling bisa dalam berbagai aplikasi, seperti dalam grafika komputer untuk memperbesar atau memperkecil objek, dalam analisis data untuk menormalkan data pada rentang tertentu, atau dalam pemrosesan gambar untuk mengubah ukuran gambar.

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks dilatasi
scaling_matrix = np.array([[2, 0], [0, 3]])  # Contoh faktor skala 2 untuk sumbu x dan 3 untuk sumbu y

# Vektor asli
original_vector = np.array([1, 3])

# Menerapkan dilatasi menggunakan matriks
scaled_vector = np.dot(scaling_matrix, original_vector)

# Plot vektor asli dan hasil dilatasi
plt.figure()
plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vektor Asli')
plt.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Hasil Dilatasi')
plt.xlim(-10, 10)  # Sesuaikan batas sumbu x sesuai dengan faktor skala yang dipilih
plt.ylim(-10, 10)  # Sesuaikan batas sumbu y sesuai dengan faktor skala yang dipilih
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Dilatasi Matriks')
plt.legend()
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Proyeksi Matriks
Proyeksi matriks adalah transformasi linier yang menghasilkan bayangan atau proyeksi dari suatu objek ke dalam sebuah subruang. Proyeksi dapat terjadi ke dalam dimensi yang lebih rendah (proyeksi ortogonal) atau ke dalam dimensi yang sama (proyeksi non-ortogonal). Proyeksi ortogonal terjadi ketika subruang yang ditentukan tegak lurus terhadap subruang lainnya. Proyeksi mengurangi dimensi vektor dengan memproyeksikannya ke suatu garis atau bidang.

Jika memiliki proyeksi vektor

$$
\begin{equation*}
v=
\begin{bmatrix}
3\\4
\end{bmatrix}
\end{equation*}
$$

ke sumbu x akan menghasilkan vektor

$$
\begin{equation*}
v'=
\begin{bmatrix}
3\\0
\end{bmatrix}
\end{equation*}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks proyeksi ke sumbu x
Px = np.array([[1, 0], [0, 0]])

# Matriks proyeksi ke sumbu y
Py = np.array([[0, 0], [0, 1]])

# Vektor asli
original_vector = np.array([1, 3])

# Menghitung hasil proyeksi ke sumbu x dan sumbu y
projected_vector_x = np.dot(Px, original_vector)
projected_vector_y = np.dot(Py, original_vector)

# Plot vektor asli dan hasil proyeksi
plt.figure()

# Vektor asli
plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vektor Asli')

# Hasil proyeksi ke sumbu x
plt.quiver(0, 0, projected_vector_x[0], projected_vector_x[1], angles='xy', scale_units='xy', scale=1, color='b', label='Proyeksi ke Sumbu X')

# Hasil proyeksi ke sumbu y
plt.quiver(0, 0, projected_vector_y[0], projected_vector_y[1], angles='xy', scale_units='xy', scale=1, color='g', label='Proyeksi ke Sumbu Y')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Proyeksi Matriks')
plt.legend()
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
