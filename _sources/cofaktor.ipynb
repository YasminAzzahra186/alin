```python
import numpy as np

# Fungsi untuk menghitung kofaktor
def cofactor(matrix, i, j):
    minor = np.delete(np.delete(matrix, i, axis=0), j, axis=1)
    cof = (-1) ** (i + j) * np.linalg.det(minor)
    return cof

# Matriks contoh
matrix = np.array([[3, 1, 3],
                   [2, 1, 2],
                   [4, 2, 1]])

# Mendapatkan ukuran matriks
rows, cols = matrix.shape

# Membuat matriks kofaktor
cofactor_matrix = np.zeros((rows, cols))

for i in range(rows):
    for j in range(cols):
        cofactor_matrix[i, j] = cofactor(matrix, i, j)

print("Matriks Kofaktor:")
print(cofactor_matrix)
```
